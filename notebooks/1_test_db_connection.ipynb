Este cuaderno verifica la conexión a la base de datos PostgreSQL utilizando las credenciales del archivo `.env`.

## IMPORTACION DE LIBRERIAS

```python
import os
import psycopg2
from dotenv import load_dotenv

load_dotenv('../.env')

print("DB_HOST:", os.getenv("DB_HOST"))
print("DB_NAME:", os.getenv("DB_NAME"))
```

## CONFIGURACION DE CONEXION

```python
try:
    conn = psycopg2.connect(
        host=os.getenv("DB_HOST"),
        database=os.getenv("DB_NAME"),
        user=os.getenv("DB_USER"),
        password=os.getenv("DB_PASSWORD"),
        port=os.getenv("DB_PORT", 5432)
    )
    print("¡Conexión exitosa a la base de datos!")
except Exception as e:
    print("Error conectando a la base de datos:", e)
```

## CONSULTAS (EJEMPLOS DE PRUEBAS)

```python
# Ejemplo de Consulta 1: Versión de PostgreSQL
if 'conn' in locals() and conn:
    try:
        cur = conn.cursor()
        cur.execute("SELECT version();")
        version = cur.fetchone()
        print("\n Versión de PostgreSQL:", version[0])
        cur.close()
    except Exception as e:
        print("Error en consulta 1:", e)
```

```python
# Ejemplo de Consulta 2: Listar algunos productos (si existen)
if 'conn' in locals() and conn:
    try:
        cur = conn.cursor()
        cur.execute("Select * from product_stocks LIMIT 5;")
        rows = cur.fetchall()
        
        print(f"\n Productos encontrados ({len(rows)}):")
        for row in rows:
            print(row)
            
        cur.close()
    except Exception as e:
        print(" No se pudo consultar la tabla 'products' (quizás no existe o está vacía):", e)
        conn.rollback() # Resetear transacción si falló
```

```python
# 5. Cerrar Conexión
if 'conn' in locals() and conn:
    conn.close()
    print("\n Conexión cerrada.")
```
