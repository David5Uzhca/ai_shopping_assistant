Vamos a probar los modelos específicos que detectamos como disponibles: `gemini-2.5-flash` y `gemini-2.0-flash`.

## IMPORTACION DE LIBERIAS

```python
import os
from google import genai
from dotenv import load_dotenv

load_dotenv('../.env')
api_key = os.getenv("GEMINI_API_KEY")
client = genai.Client(api_key=api_key)
print("Cliente inicializado.")
```

## TESTING DE MODELOS DISPONIBLES

```python
def probar_modelo(model_name):
    print(f"\n Probando modelo: {model_name}...")
    try:
        response = client.models.generate_content(
            model=model_name,
            contents="Responde solo con la palabra: FUNCIONA"
        )
        print(f"ÉXITO con {model_name}: {response.text}")
        return True
    except Exception as e:
        print(f"ERROR con {model_name}: {e}")
        return False

# -----------------------------------------------------
# PRUEBA 1: Modelo Principal (gemini-2.5-flash)
probar_modelo("gemini-2.5-flash")

# PRUEBA 2: Modelo de Respaldo (gemini-2.0-flash)
probar_modelo("gemini-2.0-flash")
```
